```python
class Node:
    def __init__(self, value, next = None, prev = None):
        self.prev = prev
        self.value = value
        self.next = next
    def __str__(self):
        return f"[Node with value {self.value}]"
        
class DoubleLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        self.length = 0

    def addElem(self, cur_node):
        if self.head == None and self.tail == None:
            cur_node.prev = self.head
            cur_node.next = self.tail
            self.head = cur_node
            self.tail = cur_node
        else:
            # Элементы добавляются в конец по умолчанию (исключение нулевой и добавление элемента в начало)
            temp = self.tail
            self.tail = cur_node
            cur_node.prev = temp
            temp.next = cur_node
        self.length += 1

    def addElemAtStart(self, cur_node):
        temp = self.head
        self.head = cur_node
        cur_node.next = temp
        temp.prev = cur_node
        self.length += 1
    
    def printList(self):
        cur_node = self.head
        while cur_node is not None:
            print(cur_node)
            cur_node = cur_node.next
    
    def reversePrintList(self):
        cur_node = self.tail
        while cur_node is not None:
            print(cur_node)
            cur_node = cur_node.prev

def reverseLinkedList(head, tail):
    p_pointer = head
    q_pointer = p_pointer.next 
    p_pointer.next = None 
    p_pointer.prev = q_pointer 
    tail = p_pointer
    while q_pointer is not None: 
        q_pointer.prev = q_pointer.next 
        q_pointer.next = p_pointer 
        p_pointer = q_pointer 
        q_pointer = q_pointer.prev
    head = p_pointer
    return head, tail

def bubbleSortLinkedList(linkedList):
    l_pointer = linkedList.head
    for i in range(linkedList.length):
        r_pointer = l_pointer.next
        for j in range(i, linkedList.length-1):
            if l_pointer.value > r_pointer.value:
                l_pointer.value, r_pointer.value = r_pointer.value, l_pointer.value
            r_pointer = r_pointer.next
        l_pointer = l_pointer.next
    return linkedList

a, b, c, g, k = Node(1), Node(2), Node(3), Node(4), Node(5)

linkedList = DoubleLinkedList()

linkedList.addElem(a)
linkedList.addElem(b)
linkedList.addElem(c)
linkedList.addElemAtStart(g)
linkedList.addElem(k)

linkedList.printList()
linkedList.head, linkedList.tail = reverseLinkedList(linkedList.head, linkedList.tail)
print("___________________")
linkedList.printList()
linkedList = bubbleSortLinkedList(linkedList)
print("___________________")
linkedList.printList()
```
